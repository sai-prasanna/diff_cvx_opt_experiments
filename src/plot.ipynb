```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
loss = torch.load(f"data/state_loss_track_e_9_i_8.pt")
print(loss)
```

```python
loss_ = [ float(x) for x in loss ]
print(loss_)
```

```python
plt.plot(list(range(len(loss_))), loss_)
plt.title("state_loss")
plt.ylabel("loss")
plt.xlabel("batch")
plt.show()
```
