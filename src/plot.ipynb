```python
loss_metrics = torch.load("data/1675863307.8865921mpc/loss_metrics.pt")
loss=loss_metrics["loss_track"]
```

```python
loss_ = [float(x) for x in loss ]
print(loss_)
```

```python
plt.plot(list(range(len(loss_))), loss_)
plt.title("loss")
plt.ylabel("loss")
plt.xlabel("batch")
plt.show()
```
